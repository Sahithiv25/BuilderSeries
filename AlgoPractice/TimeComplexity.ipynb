### How to calculate time complexity of any given algorithm

#### 5 rules of Time complexity

Rule - 1 : Loops take O(n)
Ex: for i in nums: # computes n times

Rule - 2 : Nested loops multiply
for i in nums: # n times
    for j in nums: # n times

So total of O(n^2) times

Rule - 3 : Sequential work - Add
do A # O(n)
do B # O(n^2)

So total O(n+n^2) which is O(n)


Rule - 4 : Drop any constants
for i in nums: # n times
    print(i)

O(2n) which is O(n)

Rule - 5: Worst case dominates
Ex: Binary search
while left <= right: # cuts input in half for every loop

So total O(log n)

#### Time complexity of common operations

Operation	                 Time
Access list element	         O(1)
Append to list	             O(1) 
Insert/pop from HEAD of list O(n)
Insert/pop from END of list	 O(1)
Search in unsorted list	     O(n)
Search in dict	             O(1)
Insert in dict	             O(1)
Sorting	                     O(n log n)
Set operations	             O(1) average
Any loop that repeatedly half O(log n)

Always check if the inner work is fixed/ constant or is it depended on outer loop i.e. n
Ex: 
for i in nums:
    for j in range(10): this loop is fixed not dependent on n
        print(i,j)

TC = O(N)

for i in nums:

def f2(arr):
    arr.sort()          # assume n = len(arr) # this takes O(n log n)
    n = len(arr)
    for i in range(n): # O(n)
        x = arr[i]
        # binary search on the rest of the array # takes O(log n)
        lo, hi = i+1, n-1
        while lo <= hi:
            mid = (lo + hi) // 2
            if arr[mid] == x:
                return True
            elif arr[mid] < x:
                lo = mid + 1
            else:
                hi = mid - 1
    return False

TC = O(n log n)

def f3(n):

    for i in range(n): # takes O(n) here
