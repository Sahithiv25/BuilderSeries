```python
# while loops
# continues as long as the condition is true
count = 0 
while(count<5):
    print(count)
    count+=1
```

```python
# break stmt
# break statement will exit the loop

for i in range(10):
    if i==5:
        break
    print(i)
```

```python
# continue
# continue statement will skip the current iteration and continues with next

for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
# nested loops
for i in range(4): # i is from 0 to 3
    for j in range(3): # j is from 0 to 2
        print(f"i:{i} and j:{j}")
```

```python
# Examples
# 1. sum of first n natural numbers using while and for loop
# for loop
sum = 0
for i in range(11): # i is from 0 to 10
    sum = sum + i
print(sum)
```

```python
# while loop
sum = 0
num = 10
count = 1 # you need count stop infinite loop
while(count<=num):
    sum+=count
    count+=1
print(sum)
```

```python
# 2. Prime nos: between 1 and 100
# prime no logic - a number that has no factors 2,3,5,7,11.... 
# it cannot be divided by any number - i.e. doesnt return 0
for n in range(1,101): # numbers from 1 to 100
    if n>1: # base case
        for i in range(2,n): # run from 2 to n-1
            if(n%i==0): # if it is divisible then break
                break
        else:
            print(n) # else print the no
```
