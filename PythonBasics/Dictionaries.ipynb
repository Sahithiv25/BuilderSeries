```python
# always use a single unique key no duplicate
stu1={"name":"A","age":12,"name":2}
print(stu1)
```

```python
# accessing dictionary elements
stu_dict = {"name":"A","age":12,"grade":2}
print(stu_dict['age'])
print(stu_dict['grade'])
```

```python
# accessing using the get() method
print(stu_dict.get('grade'))
print(stu_dict.get('last_name'))
print(stu_dict.get('last_name',"Not Available"))
```

```python
# Modifying the dictionary ele
stu_dict["age"] = 14 # update old key
print(stu_dict)
stu_dict["address"]="11 abc dr" # created new key
print(stu_dict)
```

```python
del stu_dict['address']
print(stu_dict)
```

```python
# check whether a key exists in a dictionary
# if matters because lookup in list is O(n) but in dict it is O(1)
d = {"a":1, "b":2}
print("a" in d)
print("c" in d)
```

```python
# default dict (from collections)
# normal dict throws error when missing key is accessed
# d={}
# d=["x"] # key error
# default dict creates a default value when a new key is accessed

from collections import defaultdict

d = defaultdict(list)
d["a"].append(20)
d["b"].append(10)
print(d)
```

```python
# dictionary methods
# 1. get all keys
keys = stu_dict.keys()
print(keys)
# 2. get all values
values = stu_dict.values()
print(values)
# 3. get all key value pairs
items=stu_dict.items()
print(items)
```

```python
# shallow copy only outer obj no inner obj
stu_copy = stu_dict
print(stu_dict)
print(stu_copy)
```

```python
# iterating over dictionaries
# loops can be used

# iterating over keys
for keys in stu_dict.keys():
    print(keys)
```

```python
# iterating over values
for values in stu_dict.values():
    print(values)
```

```python
# iterating over key value pairs
for key,value in stu_dict.items():
    print(f"{key}:{value}")
```

```python
# nested dictionaries
students={
    "stu1":{"name":"A","age":12},
    "stu2":{"name":"B","age":14}
}
print(students)
```

```python
# accessing nested dictionary elements
print(students["stu1"]["name"])
print(students["stu2"]['age'])
```

```python
students.items()
```

```python
# iterating over nested dictionaries
for stu_id, stu_info in students.items():
    print(f"{stu_id}:{stu_info}")
    for key,valye in stu_info.items():
        print(f"{key}:{value}")
```

```python
# dictionary comprehension
sq={x:x**2 for x in range(5)}
print(sq)
```

```python
# condition dictionary comprehension
even={x:x**2 for x in range(10) if x%2==0}
print(even)
```

```python
# Examples

# Count frequency of elements in list
num=[1,2,2,3,4,3,4,5,4,4,5,3,3]
freq={}
for i in num: # iterate through the list
    if i in freq: # iterate through the dict
        freq[i]+=1 # if present then increase the count
    else:
        freq[i]=1 # else 1
print(freq)
```

```python
# merge 2 dictionaries into 1
dict1={"a":1,"b":2}
dict2={"c":3,"b":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```

```python
# count freq of characters in a string
dict_str = "banana"
freq={}
for s in dict_str:
    if s in freq:
        freq[s]+=1
    else:
        freq[s]=1
print(freq)    
```

```python
# reverse a dictionary
my_dict = {"a":1, "b":2, "c":3}
reverse_dict = {v:k for k,v in my_dict.items()}
print(reverse_dict)
    
```

```python
# merge 2 dicts and keep the sum
d1 = {"a":2, "b":3}
d2 = {"b":4, "c":1}

merge_with_sum = {}
for key in set(d1) | set(d2): # union of keys
    merge_with_sum[key] = d1.get(key,0) + d2.get(key,0)
print(merge_with_sum)
```

```python
# group students by grade
stu_grades = [("A",1),("B",1),("C",2),("D",1)]
groups={}
for key,value in stu_grades: # pick the name and grafe
    if value not in groups: # if grade is not in the new dict
        groups[value] = []
    groups[value].append(key) # add value to the group based on name(key)
print(groups)


    

```
